## ✅ What is a Population?

A **population** is the entire group you want to study.

In data science terms:

* All customers of a telecom company
* All transactions in a database
* All users of a mobile app
* Entire production batch in manufacturing

**🔎 Example**

Suppose you are working on **customer churn prediction** in a telecom company.

* Total customers = 10,00,000
* You want to know the **average monthly spend**

The 10,00,000 customers = **Population**

We denote:

* Population size = $N$
* Population mean = $\mu$
* Population standard deviation =  $\sigma$



## ✅ What is a Sample?

A **sample** is a subset of the population used to estimate population characteristics.

Why?

Because:

* Collecting full population data is expensive
* Sometimes impossible
* Computationally heavy

So we take:

Example:

* Randomly select 2,000 customers from 10,00,000

That 2,000 = **Sample**

We denote:

* Sample size = $n$
* Sample mean = $\bar{x}$
* Sample standard deviation = $s$

**🎯 Why This Matters in Data Science**

In real ML systems:

* We train models on **sample data**
* We estimate parameters from samples
* We infer about unseen population

Example:

You evaluate model performance on:

* 20% validation data

That validation data = sample
But you want performance on **all future users** = population

# 2️⃣ Mathematical Formulas

## Population Mean

$$\mu = \frac{1}{N} \sum_{i=1}^{N} X_i$$

**Explanation:**

* ( $\mu$ ) = Population mean
* ( $N$) = Total population size
* ( $$X_i$ ) = Each data point




## Sample Mean

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} X_i$$

**Explanation**:

* ( $\bar{x}$ ) = Sample mean
* ( $n$ ) = Sample size
* ( $X_i$ ) = Each sample observation





## Population Variance

$$\sigma^2 = \frac{1}{N} \sum_{i=1}^{N}(X_i - \mu)^2$$





## Sample Variance (Important Difference)

$$s^2 = \frac{1}{n-1} \sum_{i=1}^{n}(X_i - \bar{x})^2$$

**⚠ Why ( n - 1 )?**

This is called **Bessel’s Correction**.

Because:

* Sample underestimates population variance
* Dividing by ( n - 1 ) corrects the bias

# 3️⃣ Dummy Dataset

You are analyzing **monthly spend of customers**.

We simulate:

* Population = 10,000 customers
* Sample = 200 customers


```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
# Step 1: Create Population (10,000 customers monthly spend)
population = np.random.normal(loc=5000, scale=800, size=10000)
```

```python

# Step 2: Take Random Sample (200 customers)
sample = np.random.choice(population, size=200, replace=False)
```

```python

# Step 3: Population Statistics
population_mean = np.mean(population)
population_std = np.std(population)
```

```python
# Step 4: Sample Statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # ddof=1 for sample std
```

```python
population_mean, sample_mean
```


# Interpretation 

| Concept | Population | Sample      |
| ------- | ---------- | ----------- |
| Symbol  | ( $\mu$ )    | ( $\bar{x}$ ) |
| Size    | ( $N$ )      | ( $n$ )       |
| Std Dev | ( $\sigma$ ) | ( $s$ )       |
| Used in | $Reality$    | $Estimation$  |




# Interview Angle

Interviewer:
"Why don’t we calculate population mean directly?"

Answer:

> In real-world systems, we rarely have access to the full population. Even if we do, it is computationally inefficient. So we estimate population parameters using samples and quantify uncertainty using statistical inference.

## What is Sampling Distribution?

A **sampling distribution** is:

> The distribution of a statistic (like mean) computed from multiple samples drawn from the same population.

Very important:

It is NOT the distribution of data.
It is the distribution of a statistic.





## Real Data Science Scenario

You are a data analyst in an e-commerce company.

Population:

* 5,00,000 customers
* Average order value = unknown

You:

* Take sample of 200 customers
* Calculate sample mean

Now imagine:

You repeat this 1,000 times.

Each time → different sample mean.

The distribution of those 1,000 sample means
= Sampling Distribution of the Mean


Why This Matters in ML

* Model evaluation is based on samples
* A/B testing uses sample metrics
* Cross-validation scores are sample estimates

Cross-validation scores themselves form a sampling distribution.


# 2️⃣ Mathematical Formula

If:

Population mean = ( $\mu$ )

Population standard deviation = ( $\sigma$ )

Sample size = ( $n$)

Then:

### Mean of Sampling Distribution:

$$\mu_{\bar{x}} = \mu $$

Meaning:

Sample mean is an unbiased estimator.





### Standard Deviation of Sampling Distribution:

$$\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$$

This is called **Standard Error**.

# 3️⃣ Standard Error (SE)

![pic](https://www.researchgate.net/publication/393195231/figure/fig1/AS:11431281525232060@1751357227336/Sampling-distribution-and-SEM.png)

[source](https://www.researchgate.net/figure/Sampling-distribution-and-SEM_fig1_393195231)



Standard Error measures:

> How much sample mean varies from sample to sample.

Formula:

$$SE = \frac{\sigma}{\sqrt{n}}$$

If population std is unknown (real life case):

$$SE = \frac{s}{\sqrt{n}}$$

Where:

* ( $s$ ) = sample standard deviation
* ( $n$ ) = sample size


**Intuition**

Large sample size → smaller SE

Small sample size → larger SE

Because:

More data → more stable estimate.





# 4️⃣ Python Simulation (Very Important)

We will:

1. Create population
2. Draw 1,000 samples
3. Store their means
4. Plot sampling distribution


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Step 1: Create population
population = np.random.normal(loc=5000, scale=800, size=100000)

population_mean = np.mean(population)
population_std = np.std(population)

# Step 2: Generate sampling distribution
sample_size = 100
num_samples = 1000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Calculate Standard Error
theoretical_se = population_std / np.sqrt(sample_size)
empirical_se = np.std(sample_means)

print("Population Mean:", population_mean)
print("Mean of Sample Means:", np.mean(sample_means))
print("Theoretical SE:", theoretical_se)
print("Empirical SE:", empirical_se)

# Step 4: Plot Sampling Distribution
plt.figure(figsize=(10,6))
sns.histplot(sample_means, kde=True)
plt.axvline(population_mean, color='red', linestyle='--', label='Population Mean')
plt.title("Sampling Distribution of the Sample Mean")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**What You Will Observe**

1. The sampling distribution is approximately normal.
2. Centered around population mean.
3. Spread equals Standard Error.
4. Much narrower than population distribution.




**Key Insight**

Population std = 800

Sample size = 100

Standard Error:

$$SE = \frac{800}{\sqrt{100}} = 80$$

Meaning:

Sample means vary around ±80 from true mean.



**Interview-Level Understanding**

Interviewer:
"What happens to standard error if sample size increases?"

Answer:

$$SE \propto \frac{1}{\sqrt{n}}$$

If sample size increases 4 times → SE reduces by half.





Interviewer:

"Why is sampling distribution important in hypothesis testing?"

Strong Answer:

> Hypothesis testing is based on how far our sample statistic lies from the population parameter measured in units of standard error.

That distance becomes:

* Z-score
* T-score




**Comparison Table**

| Concept | Population Distribution | Sampling Distribution |
| ------- | ----------------------- | --------------------- |
| Spread  | ( $\sigma$ )              | ( $\sigma$ / \sqrt{n}$ ) |
| Mean    | ( $\mu$ )                 | ( $\mu$ )               |
| Shape   | Any shape               | Normal (if n large)   |
| Use     | Reality                 | Inference             |



**What We Understood**

1. Sampling distribution = distribution of statistic
2. Mean stays same
3. Spread reduces
4. Spread = Standard Error
5. Hypothesis testing uses SE
6. Larger sample → more confidence


- Population distribution = variability in data
- Sampling distribution = variability in estimate

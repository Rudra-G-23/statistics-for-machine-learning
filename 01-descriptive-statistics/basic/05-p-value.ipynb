# PART 1 — Type I and Type II Errors


## 1. Theory

Whenever we test a hypothesis, two things are possible in reality:

* Null hypothesis is actually true
* Null hypothesis is actually false

But our decision has only two options:

* Reject H₀
* Fail to reject H₀

This creates four possibilities.

Truth vs Decision table:

Reality: H₀ True

* Reject H₀ → Type I Error
* Fail to Reject H₀ → Correct Decision

Reality: H₀ False

* Reject H₀ → Correct Decision
* Fail to Reject H₀ → Type II Error

## Type I Error (False Positive)

Definition:

Rejecting the null hypothesis when it is actually true.

Probability of Type I error:

$$ 
P(\text{Reject } H_0 \mid H_0 \text{ true}) = \alpha
$$


````python
import numpy as np
from scipy import stats

np.random.seed(42)

population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05

num_experiments = 10000
false_rejections = 0

for _ in range(num_experiments):
    sample = np.random.normal(loc=population_mean,
                              scale=population_std,
                              size=sample_size)
    
    sample_mean = np.mean(sample)
    
    # standard error
    se = population_std / np.sqrt(sample_size)
    
    # test statistic
    test_stat = (sample_mean - population_mean) / se
    
    # two-sided p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(test_stat)))
    
    if p_value < alpha:
        false_rejections += 1

type1_error_rate = false_rejections / num_experiments
print("Estimated Type I Error Rate:", type1_error_rate)
````

## Type II Error (False Negative)

Definition:

Failing to reject the null hypothesis when it is actually false.

Probability of Type II error:

$$
P(\text{Fail to Reject } H_0 \mid H_0 \text{ false}) = \beta
$$


````python
true_mean = 105
type2_failures = 0

for _ in range(num_experiments):
    sample = np.random.normal(loc=true_mean,
                              scale=population_std,
                              size=sample_size)
    
    sample_mean = np.mean(sample)
    se = population_std / np.sqrt(sample_size)
    test_stat = (sample_mean - population_mean) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(test_stat)))
    
    if p_value >= alpha:
        type2_failures += 1

type2_error_rate = type2_failures / num_experiments
power = 1 - type2_error_rate

print("Estimated Type II Error Rate:", type2_error_rate)
print("Estimated Power:", power)
````

## Power of a Test:


$$\text{Power} = 1 - \beta
 $$ 

Meaning: Probability of correctly detecting a real effect.

## Real Data Science Example

Suppose you are running an A/B experiment.

H₀: New feature does not change conversion rate.

Type I Error:
You conclude feature improved conversion, but actually it did not.

Business impact:
You deploy useless feature.

Type II Error:
Feature actually improves conversion, but you conclude it does not.

Business impact:
You miss revenue opportunity.

# PART 2 — Significance Level (α)

---

Definition:

Significance level is the maximum probability of committing a Type I error that we are willing to tolerate.

Common value:

[
\alpha = 0.05
]

Meaning:

We accept a 5% risk of false positive.

Important:

Alpha is chosen before looking at data.

Lower alpha:

* Harder to reject H₀
* Fewer false positives
* More false negatives

---

PART 3 — P-value (Concept)

---

Definition:

P-value is the probability of observing data at least as extreme as what we observed, assuming the null hypothesis is true.

Mathematically:

[
p = P(\text{Data as extreme as observed} \mid H_0 \text{ true})
]

Decision rule:

If

[
p < \alpha
]

Reject H₀.

Otherwise, fail to reject H₀.

Important:

P-value is not:

* Probability that H₀ is true
* Probability that results happened by chance

It is conditional on H₀ being true.

EXAMPLE 1 — Simulating Type I Error

---

We simulate a situation where H₀ is true.

Population mean = 100
We repeatedly sample and check how often we incorrectly reject.

Code:

```python
import numpy as np
from scipy import stats

np.random.seed(42)

population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05

num_experiments = 10000
false_rejections = 0

for _ in range(num_experiments):
    sample = np.random.normal(loc=population_mean,
                              scale=population_std,
                              size=sample_size)
    
    sample_mean = np.mean(sample)
    
    # standard error
    se = population_std / np.sqrt(sample_size)
    
    # test statistic
    test_stat = (sample_mean - population_mean) / se
    
    # two-sided p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(test_stat)))
    
    if p_value < alpha:
        false_rejections += 1

type1_error_rate = false_rejections / num_experiments
print("Estimated Type I Error Rate:", type1_error_rate)
```



Expected output ≈ 0.05

This proves:

Alpha = probability of Type I error.




EXAMPLE 2 — Simulating Type II Error



````python
true_mean = 105
type2_failures = 0

for _ in range(num_experiments):
    sample = np.random.normal(loc=true_mean,
                              scale=population_std,
                              size=sample_size)
    
    sample_mean = np.mean(sample)
    se = population_std / np.sqrt(sample_size)
    test_stat = (sample_mean - population_mean) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(test_stat)))
    
    if p_value >= alpha:
        type2_failures += 1

type2_error_rate = type2_failures / num_experiments
power = 1 - type2_error_rate

print("Estimated Type II Error Rate:", type2_error_rate)
print("Estimated Power:", power)
````


You will observe:

* Type II error depends on effect size.
* Larger difference → lower Type II error.
* Larger sample size → lower Type II error.

EXAMPLE 3 — Understanding P-value Directly



````python
sample = np.random.normal(loc=100,
                          scale=15,
                          size=30)

sample_mean = np.mean(sample)
se = 15 / np.sqrt(30)
test_stat = (sample_mean - 100) / se
p_value = 2 * (1 - stats.norm.cdf(abs(test_stat)))

print("Sample Mean:", sample_mean)
print("Test Statistic:", test_stat)
print("P-value:", p_value)
````


Interpretation:

If p_value = 0.03 and alpha = 0.05:

0.03 < 0.05
We reject H₀.

Meaning:

There is strong evidence against the null hypothesis.





# Deep Understanding Summary

Type I Error:
False alarm. Controlled by alpha.

Type II Error:
Missed detection. Reduced by larger sample size.

Alpha:
Risk threshold chosen before experiment.

P-value:
Evidence measure against null hypothesis.

Relationship:

Smaller alpha:

* Fewer Type I errors
* More Type II errors

Larger sample:

* Reduces Type II error
* Increases power





# Interview-Level Understanding

If interviewer asks:

Why can’t we reduce both Type I and Type II errors to zero?

Answer:

Because reducing Type I error by lowering alpha makes rejection harder, which increases Type II error unless we increase sample size.

Trade-off exists.

# 1. Why Do We Need Hypotheses?

In data science and business, we often want to check:

* Did a new feature improve conversion?
* Did churn rate decrease?
* Is average revenue different?
* Did a model improve performance?

We cannot rely on intuition.
We need a structured way to make decisions under uncertainty.

Hypothesis testing provides that structure.

The first step is defining two competing statements:

* Null Hypothesis
* Alternative Hypothesis


# 2. Null Hypothesis (H₀)

Definition:

The Null Hypothesis is the default assumption that there is no effect, no difference, or no change.

It represents the current belief or status quo.

Mathematically, it usually contains equality.

Examples:

If population mean is ( $\mu$ ) and claimed value is ( $\mu_0$ ), then:

$$ H_0: \mu = \mu_0 $$

Key characteristics:

* Assumes nothing has changed.
* Assumes observed differences are due to random sampling.
* It is the statement we try to test against.

Important interview point:

We never try to prove the null hypothesis.
We test whether we have enough evidence to reject it.

# 3. Alternative Hypothesis (H₁ or Hₐ)

Definition:

The Alternative Hypothesis is the claim that there is an effect, difference, or change.

It contradicts the null hypothesis.

Examples:

Two-sided (difference exists):

$$ H_1: \mu \ne \mu_0 $$

Right-tailed (increase):

$$  H_1: \mu > \mu_0 $$

Left-tailed (decrease):

$$
H_1: \mu < \mu_0
$$

Key characteristics:

* Represents what we are trying to find evidence for.
* It is accepted only if the null hypothesis is rejected.


# 4. Real Data Science Scenario

**Scenario 1: Feature Launch**

You work in an e-commerce company.

Current average conversion rate = 5%.

You launch a new UI design.

Null hypothesis:

$$ H_0: \text{New conversion rate} = 5% $$

Alternative hypothesis:

$$ H_1: \text{New conversion rate} \ne 5% $$

Meaning:

If data shows strong evidence of change, we reject H₀.



**Scenario 2: Model Improvement**

You improved a churn prediction model.

Old model accuracy = 85%.

You want to test if new model is better.

Null hypothesis:

$$  H_0: \text{New accuracy} = 85% $$

Alternative hypothesis:

$$ H_1: \text{New accuracy} > 85% $$

If evidence supports improvement, we reject H₀.

# 5. Types of Alternative Hypothesis

Two-sided:

$$ H_1: \mu \ne \mu_0 $$

Used when any difference matters.

One-sided (Right-tailed):

$$ H_1: \mu > \mu_0 $$

Used when only improvement matters.

One-sided (Left-tailed): 

$$
H_1: \mu < \mu_0
$$

Used when only decrease matters.



**Interview insight:**

Choosing one-sided after seeing data is incorrect.
It must be defined before analysis.



# 6. Logical Structure

Hypothesis testing works like a court system.

* Null hypothesis = “Innocent”
* Alternative hypothesis = “Guilty”

We assume innocence until strong evidence proves guilt.

Similarly:

We assume no effect until strong evidence proves otherwise.



# 7. Common Mistakes 

Mistake 1:
Thinking we “accept” the null hypothesis.

Correct statement:
We fail to reject the null hypothesis.

Mistake 2:
Confusing hypothesis with conclusion.

Hypotheses are defined before looking at data.

Mistake 3:
Writing alternative hypothesis without clear direction.



# 8. Conceptual Summary

Null Hypothesis:

* Default assumption
* No change
* Contains equality
* Assumed true initially

Alternative Hypothesis:

* Opposes null
* Represents effect or difference
* Accepted only if null is rejected
